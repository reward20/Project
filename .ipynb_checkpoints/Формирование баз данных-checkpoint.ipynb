```python
from pathlib import Path
import pandas as pd
from shutil import copy
from pandas.io.excel import ExcelWriter
import time
```

```python
class BD_Maker():
    
    def _raise_decorate(def_fun):
        #Обработчик ошибок
        def raise_check(self):
            try:
                return def_fun(self)
            except Exception as err:
                print("!!!Ошибка!!!")
                print(err)
        return raise_check
    
    
    
    def __init__(self):
        
        #Таблица Маршрутные листы
        self.df_ml = None 
        self.df_detail = None
        self.df_operation = None
        self.df_sklad = None
        self.df_order = None
        
        #Дата фреймы с которыми класс будет работать
        self.MLEXCEL = None
        #Дата фрейм для МП 
        self.MLEXCEL_MP = None
        #Дата фрейм для УЗ
        self.MLEXCEL_UZ = None
        

        
        self.__url_Base = None
        self.__url_MLEXCEL = None
        self.__url_MLEXCEL_MP = None
        self.__url_MLEXCEL_UZ = None
        self.__url_dos_file = None
        self.__url_csv_file = None
        self.__url_copy_dos_file = None
    
    
    @property
    def url_Base(self):
        return str(self.__url_Base)
    
    @url_Base.setter
    def url_Base(self,url):
        self.__url_Base = Path(url)
    
    @property
    def url_MLEXCEL_MP(self):
        return str(self.__url_MLEXCEL_MP)
    
    @url_MLEXCEL_MP.setter
    def url_MLEXCEL_MP(self, url):
        self.__url_MLEXCEL_MP = Path(url)
        
    @property
    def url_MLEXCEL_UZ(self):
        return str(self.__url_MLEXCEL_UZ)
    
    @url_MLEXCEL_UZ.setter
    def url_MLEXCEL_UZ(self, url):
        self.__url_MLEXCEL_UZ = Path(url)
    
    @property
    def url_dos_file(self):
        return str(self.__url_dos_file)
    
    @url_dos_file.setter
    def url_dos_file(self, url):
        self.__url_dos_file = Path(url)
    
    @property
    def url_dos_file(self):
        return str(self.__url_dos_file)
    
    @url_dos_file.setter
    def url_dos_file(self, url):
        self.__url_dos_file = Path(url)

        
    @property
    def url_MLEXCEL(self):
        return str(self.__url_MLEXCEL)
    
    @url_MLEXCEL.setter
    def url_MLEXCEL(self, url):
        self.__url_MLEXCEL = Path(url)
        
        
    @property
    def url_csv_file(self):
        return str(self.__url_csv_file)
    
    @url_csv_file.setter
    def url_csv_file(self, url):
        self.__url_csv_file = Path(url)
        
        
    @property
    def url_copy_dos_file(self):
        return str(self.__url_copy_dos_file)
    
    @url_copy_dos_file.setter
    def url_copy_dos_file(self, url):
        self.__url_copy_dos_file = Path(url)
    
    @_raise_decorate
    def start_make_bd(self):
        #Словарь с полученными путями файлов MLEXCEL, MLEXCELO, SKLAD 
        
        #Если не указан путь с MLEXCEL, MLEXCELO, SKLAD 
        if self.__url_dos_file is None:
            raise Exception("Не указан путь с файлами данных")    
        
        self.__dict_path = self.__check_url_dos_file()
        self.__read_bd_file()
        
        #Если не указан ни один путь то закончить выполнение
        if self.__url_Base is None and \
            self.__url_MLEXCEL_MP is None and \
            self.__url_MLEXCEL is None and \
            self.__url_MLEXCEL_UZ is None and \
            self.__url_csv_file is None:
            raise Exception("Не указана ни одна ссылка для выходных данных")
        
        """
            Формирование таблицы общей базы
            Запись данных в csv формат
        """
        if not (self.__url_Base is None and \
            self.__url_csv_file is None):
            self.__create_Base()
            if not(self.__url_Base is None):
                if  self.__url_Base.is_dir():
                    self.__write_excel_Base()
                else:
                    print(f"Нет доступа к {self.url_Base} Base")

            if not(self.__url_csv_file is None):
                if self.__url_csv_file.is_dir():
                    self.__write_csv_file()
                else:
                    print(f"Нет доступа к {self.url_csv_file} CSV")   
        
        """Формирование файла для МП"""
        if not self.__url_MLEXCEL_MP is None:
            if self.__url_MLEXCEL_MP.is_dir():
                self.__create_MLEXCEL_MP()
                self.__write_excel_MLEXCEL_MP()
            else:
                print(f"Нет доступа к папке {self.url_MLEXCEL_MP} МП ")
        
        """Формарование файла для УЗ"""
        if not self.__url_MLEXCEL_UZ is None:
            if self.__url_MLEXCEL_UZ.is_dir():
                self.__create_MLEXCEL_UZ()
                self.__write_excel_MLEXCEL_UZ()
            else:
                print(f"Нет доступа к {self.url_MLEXCEL_UZ} UZ")
        
        """Формирование файла для ПДО"""
        if not self.__url_MLEXCEL is None:
            if self.__url_MLEXCEL.is_dir():
                self.__write_excel_MLEXCEL()
            else:
                print(f"Нет доступа к папке {self.url_MLEXCEL} ПДО")
        
        
    def __check_url_dos_file(self) -> dict:
        """Проверяет доступ к пути и наличие необходимых файлов"""
        #Если путь к папку не найден
        if not self.__url_dos_file.is_dir():
            raise(Exception("Нет доступа к папке с файлами дос"))
    
        check_file_name = ["MLEXCEL", "MLEXCELO", "SKLAD"]
        dict_path = dict()
        for file in self.__url_dos_file.iterdir():
            if file.name in check_file_name:
                check_file_name.remove(file.name)
                dict_path.update({file.name:file})
            if len(check_file_name) == 0:
                if not self.__url_copy_dos_file is None:
                    if self.__url_copy_dos_file.is_dir():
                        for path in dict_path.values():
                            copy(path,self.__url_copy_dos_file)
                    else:
                        raise(Exception("Нет доступа к папке резервного копирования DOS"))
                return dict_path
        raise(Exception("Не найдены MLEXCEL, MLEXCELO, SKLAD"))        
        
    def __read_bd_file(self):
        #Cчитывает файлы MLEXCEL, MLEXCELO, SKLAD в таблицы
               
        def MLEXCEL():
            #Чтение файла MLEXCEL 
            names = ["Маршрутный лист","Заказ","№ детали","Название детали","Количество","Норма расхода на ед","Норм/часы","Материал","Масса детали","Профиль заготовки","Профиль","Дата распечатки", "Деталей из заготовки"]
            self.MLEXCEL = pd.read_csv(self.__dict_path["MLEXCEL"] ,delimiter="^",encoding = "cp866",decimal=",",names = names, keep_default_na = False,dtype = {"Маршрутный лист":str})

        def MLEXCELO():
            #Чтение файла MLEXCELO
            with open(self.__dict_path["MLEXCELO"],"r",encoding = "cp866") as file:
                #Возвращаемый словарь с полученным тех процессом
                df_dict ={
                          "Заказ":[],
                          "Маршрутный лист":[],
                          "№ детали":[],
                          "Название детали":[],
                          "Размер партии": [],
                          "Номер операции":[],
                          "Название операции":[],
                          "Тпз":[],
                          "Тшт":[],
                          "Тштк": []
                        }
                #Переменная для хранения названия детали
                mr_data = {}
                for line in file:
                    if line[0] == "@":
                        #Обработка первой строки
                        data_list = line[1:].split("^")
                        mr_data["Маршрутный лист"] = data_list[0]
                        mr_data["Заказ"] = data_list[1]
                        mr_data["№ детали"] = data_list[2]
                        mr_data["Название детали"] = data_list[3]
                        mr_data["Размер партии"] = data_list[4]
                        
                    elif line[0] == "N":
                        #Обработка строки операции с добавлением к ней данных из строки с мр
                        
                        oper_row = line.split(":")
                        #Операции которые < 100 считаются битыми
                        if not ((oper_row[0][1:].isnumeric()) and int(oper_row[0][1:]) > 99):
                            continue
                        oper_line = oper_row[1].split("^")   
                        df_dict["Заказ"].append(mr_data["Заказ"])
                        df_dict["Маршрутный лист"].append(mr_data["Маршрутный лист"])
                        df_dict["№ детали"].append(mr_data["№ детали"])
                        df_dict["Название детали"].append(mr_data["Название детали"])
                        df_dict["Размер партии"].append(mr_data["Размер партии"])
                        df_dict["Номер операции"].append(oper_row[0][1:])
                        df_dict["Название операции"].append(oper_line[0])
                        df_dict["Тпз"].append(float(oper_line[1].replace(",",".")))
                        df_dict["Тшт"].append(float(oper_line[2].replace(",",".")))
                        df_dict["Тштк"].append(float(oper_line[3].replace(",",".")))         
            self.df_operation = pd.DataFrame(df_dict)            


        def SKLAD():
            #Считывает данные о сдаче в таблицу df_sklad
            names = ["№ детали","Маршрутный лист","Тип сдачи","Количество","Дата сдачи"]
            self.df_sklad = pd.read_csv(self.__dict_path["SKLAD"],sep = "^",header = None, names = names, decimal = ",",encoding = "cp866",dtype = {"Маршрутный лист":str})

        MLEXCEL()
        MLEXCELO()
        SKLAD()
    
    
    """
        Функции формирования таблиц 
    """
    
    def __create_Base(self):      
        #Получение таблицы Маршрутные листы
        self.df_ml = self.MLEXCEL.merge(self.df_sklad[["Маршрутный лист","№ детали","Тип сдачи","Дата сдачи"]], on = ["Маршрутный лист","№ детали"], how = "left").fillna("")
        del self.df_sklad
        oper_ml_d = self.df_operation.groupby("Маршрутный лист")["Тштк"].sum().to_frame().reset_index()
        self.df_ml = self.df_ml.merge(oper_ml_d,on = "Маршрутный лист", how = "left")
        self.df_ml["Тштк"] = self.df_ml["Тштк"].round(4)
        del oper_ml_d
        self.df_ml.drop("Норм/часы",axis = 1,inplace = True)
        self.df_ml.rename(columns = {"Тштк":"Норм/часы"},inplace = True)

        #Получение таблицы Операции
        self.df_operation = self.df_operation[["№ детали","Номер операции","Название операции","Тпз","Тшт"]]  
        self.df_operation.drop_duplicates(subset = ["№ детали","Номер операции"], ignore_index = True, inplace = True)
        self.df_operation.set_index(["№ детали","Номер операции"], inplace = True)

        #Получение таблицы Детали
        names = ["№ детали","Название детали","Профиль заготовки","Материал","Деталей из заготовки","Норма расхода на ед","Масса детали"]
        self.df_detail = self.MLEXCEL[names].copy()
        self.df_detail.drop_duplicates(subset = "№ детали", ignore_index= True, inplace = True)
        
        #Получение таблицы Заказ
        self.df_order = pd.DataFrame(self.df_ml["Заказ"].sort_values().unique(), columns = ["Заказ"])
        self.df_order["Кол. наименований"] = self.df_order.apply(lambda tab: self.df_ml[self.df_ml["Заказ"] ==  tab["Заказ"]]["№ детали"].unique().size,axis = 1)
        self.df_order["Кол. деталей"] = self.df_order.apply(lambda tab: self.df_ml[self.df_ml["Заказ"] ==  tab["Заказ"]]["Количество"].sum(),axis = 1)
        self.df_order["Норм/часы"] = self.df_order.apply(lambda tab: self.df_ml[self.df_ml["Заказ"] ==  tab["Заказ"]]["Норм/часы"].sum(),axis = 1)
        self.df_order["Норм/часы"] = self.df_order["Норм/часы"].round(3)
        table = self.df_ml[self.df_ml["Тип сдачи"] ==  "D"]
        self.df_order["Выполнено %"] = self.df_order.apply(lambda tab: round((table[table["Заказ"] ==  tab["Заказ"]]["Норм/часы"].sum())/tab["Норм/часы"]*100,2),axis = 1)
        del table
        
    def __create_MLEXCEL_MP(self):
        #Получение таблицы MLEXCEL_MP
        name = ["Маршрутный лист","Заказ","№ детали","Название детали","Количество","Норма расхода на ед","Норм/часы","Материал","Масса детали"]
        self.MLEXCEL_MP = self.MLEXCEL[name].copy()
        self.MLEXCEL_MP["Маршрутный лист"] = self.MLEXCEL_MP["Маршрутный лист"].astype(int)
        
    def __create_MLEXCEL_UZ(self):
        #Получение таблицы MLEXCEL_UZ
        name = ["Маршрутный лист","Заказ","№ детали","Название детали","Материал","Профиль заготовки","Профиль","Норма расхода на ед","Деталей из заготовки","Количество"]
        self.MLEXCEL_UZ = self.MLEXCEL[name].copy()
        self.MLEXCEL_UZ.rename(columns = {"Количество": "Размер партии"}, inplace = True)
        self.MLEXCEL_UZ["Расход на партию"] = self.MLEXCEL_UZ.apply(lambda row: row["Размер партии"] * row["Норма расхода на ед"], axis = 1)
        self.MLEXCEL_UZ = self.MLEXCEL_UZ[~self.MLEXCEL_UZ["Заказ"].isin(["31799"])]
        self.MLEXCEL_UZ["Деталей из заготовки"] = self.MLEXCEL_UZ["Деталей из заготовки"].astype(int)
        del name

    """
        Функции записи таблиц в эксель
    """

    def __write_excel_Base(self):
        with ExcelWriter(self.__url_Base/"Базы.xlsx",mode = "w",engine = "xlsxwriter") as excel_f:
            self.df_order.to_excel(excel_f,sheet_name = "Заказы",freeze_panes= (1,0),index = False)
            self.df_ml.to_excel(excel_f,sheet_name = "Мр листы",freeze_panes= (1,0),index = False)
            self.df_operation.to_excel(excel_f,sheet_name = "Операции",freeze_panes= (1,0))
            self.df_detail.to_excel(excel_f,sheet_name = "Детали",freeze_panes= (1,0),index = False)

    def __write_excel_MLEXCEL_MP(self):
        self.MLEXCEL_MP.to_excel(self.__url_MLEXCEL_MP/"MLEXCEL.xlsx",index = False, sheet_name ="MLEXCEL",freeze_panes= (1,0))
    
    def __write_excel_MLEXCEL_UZ(self):
        self.MLEXCEL_UZ.to_excel(self.__url_MLEXCEL_UZ/"ПОТРЕБНОСТЬ МЕТАЛЛА НА ЗАКАЗЫ.xlsx",index = False, sheet_name ="Потребность",freeze_panes= (1,0))
        
    
    def __write_excel_MLEXCEL(self):
        self.MLEXCEL.to_excel(self.__url_MLEXCEL/"MLEXCEL_PDO.xlsx",index = False, sheet_name = "MLEXCEL",freeze_panes= (1,0))
        
       
        
    def __write_csv_file(self):
        self.df_operation.to_csv(self.__url_csv_file/"Операции.csv", sep = "^", encoding = "cp1251")
        self.df_ml.to_csv(self.__url_csv_file/"Маршрутные листы.csv", sep = "^", encoding = "cp1251",index = False)
        self.df_order.to_csv(self.__url_csv_file/"Заказы.csv", sep = "^", encoding = "cp1251", index = False)
        self.df_detail.to_csv(self.__url_csv_file/"Детали.csv", sep = "^", encoding = "cp1251", index = False)

        
    
    
```

```python
#Отладочный запуск
timer = time.time()
v = BD_Maker()

v.url_dos_file = r".\DOS_File"
v.url_csv_file = r".\LC"
v.url_MLEXCEL_MP = r".\LC"
v.url_MLEXCEL_UZ = r".\LC"
v.url_Base = r".\LC"
v.url_MLEXCEL = r".\LC"
v.url_copy_dos_file = r".\LC"

v.start_make_bd()
print(time.time() - timer)
```
